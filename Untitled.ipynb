```python
import numpy as np
import matplotlib.pyplot as plt

def inf_func(N):
    t = 0
    n_list = np.zeros(N+1)
    for n in range(N+1):
        n_list[n] = t
        t = t + n
    return n_list

def inf_exp(N):
    t = 0
    n_list = np.zeros(N+1)
    for n in range(N+1):
        n_list[n] = t
        t = t + n*np.exp(-n/N)
    return n_list

def inf_cos(N):
    t = 0
    n_list = np.zeros(N+1)
    for n in range(N+1):
        n_list[n] = t
        t = t + n*(np.exp(-n/N)*np.cos(n/N))
    return n_list
```
